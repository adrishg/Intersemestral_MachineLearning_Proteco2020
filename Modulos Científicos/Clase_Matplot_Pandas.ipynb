```
import matplotlib.pyplot as plt
import numpy as np
```

```
x=np.arange(0.,11.,1)
y_lineal= x
y_cuadrada= x**2
plt.plot(x,y_lineal, '+r', label='Lineal', markersize=5)
plt.plot(x,y_cuadrada, '--g', label="Cuadrada", linewidth=7)
plt.title('Gr√°fica lineal')
plt.xlabel('Eje x')
plt.ylabel('Eje y')
plt.legend()
plt.grid(True)
plt.show()
```

```
x=np.arange(0.,10., 0.2)
seno=np.sin(x)
coseno=np.cos(x)

plt.plot(x,seno, label="Seno")
plt.plot(x,coseno, label="Coseno")
plt.title('Trigonom√©tricas')

plt.legend()
plt.grid(True)
plt.show()
```

```
fig=plt.figure()
ax=fig.add_subplot()

lineas=np.arange(0,11,1)

ax.set_xticks(lineas)
ax.set_yticks(lineas)
ax.grid(True, which='both')
plt.show()
```

```
monedas = {'unPeso':8, 'dosPesos':3, 'cincoPesos':6, 'diezPesos':2}
nombreMonedas= list(monedas.keys())
numeroMonedas=list(monedas.values())
```

```
plt.bar(nombreMonedas, numeroMonedas, color='cyan', edgecolor='black', linewidth='5')
plt.title('Monedas (barras)')
plt.show()
```

```
plt.scatter(nombreMonedas, numeroMonedas, s=15,color='green', marker='o')
plt.title('Monedas (Dispersion)')
plt.show()
```

```
plt.pie(numeroMonedas, labels=nombreMonedas, colors=['blue','cyan','yellow','green'], autopct='%.4f%%', startangle=90)
plt.title('Monedas (pastel)')
plt.show()
```

```
N=50
x=np.random.rand(N)
y=np.random.rand(N)
colors = np.random.rand(N)
area = (30*np.random.rand(N))**2

plt.scatter(x,y, s=area, c=colors, alpha=0.5)
plt.show()
```

```
figura, (grafica1, grafica2)= plt.subplots(1,2, sharey=True, figsize=(10,4))
grafica1.bar(nombreMonedas, numeroMonedas)
grafica2.scatter(nombreMonedas, numeroMonedas)
plt.show()
```

```
from mpl_toolkits.mplot3d import Axes3D

fig= plt.figure()
ax=Axes3D(fig)

N=100
x = np.random.rand(N)
y = np.random.rand(N)
z = np.random.rand(N)

ax.scatter(x,y,z, color='red')
ax.set_xlabel('Eje x')
ax.set_ylabel('Eje y')
ax.set_zlabel('Eje z')
plt.show()
```

```
def z(x,y):
  return np.cos(np.sqrt(x**2+y**2))
```

```
from mpl_toolkits.mplot3d import axes3d
import matplotlib.pyplot as plt

fig = plt.figure()
ax=Axes3D(fig)

x= np.linspace(-4,4,50)
y= np.linspace(-4,4,50)
X,Y=np.meshgrid(x,y)

ax.plot_surface(X,Y,z(X,Y), color='green')
plt.show()

```
