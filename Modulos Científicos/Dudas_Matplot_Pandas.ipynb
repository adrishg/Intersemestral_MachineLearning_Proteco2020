```python
import matplotlib.pyplot as plt
import numpy as np
from mpl_toolkits.mplot3d import Axes3D
```

## Duda sobre como cambiar el tamaño del grid

```python
#Datos a utilizar
x=np.arange(0.,10., 0.2)
seno=np.sin(x)
coseno=np.cos(x)

#Para modificar el grid debemos tratar a la gráfica como lo hicimos en 3D
#Es decir debemos crear primero el escenario y luego graficar
fig=plt.figure()

#Con add_subplot(1,1,1) le decimos que en nuestro contenedor vacío le vamos a añadir una gráfica
ax=fig.add_subplot(1,1,1)

#Creamos un arreglo en el que le decimos qué tan distanciadas queremos las lineas del grid
lineas=np.arange(0,11,1.5)

#Con set_xticks le decimos al escenario que en el eje x ponga las lineas del grid a la separacion indicada por la variable lineas
#Con minor=True le decimos que las lineas no tiene que ir numeradas
ax.set_xticks(lineas, minor=True)
#Lo mismo aplica para set_yticks
ax.set_yticks(lineas, minor=True)
#Con grid which 'both' le decimos al escenario que aplique los cambios a ambos ejes
ax.grid(which='both')

ax.plot(x,seno)
ax.plot(x,coseno)

#Mostramos la gráfica
plt.show()
```

## Duda sobre rotar una gráfica 3D 

```python
def z(x,y):
    return np.cos(np.sqrt(x**2+y**2))
```

```python
fig = plt.figure()
ax=Axes3D(fig)

x= np.linspace(-4,4,50)
y= np.linspace(-4,4,50)
X,Y=np.meshgrid(x,y)

ax.plot_surface(X,Y,z(X,Y), color='green')

#Para girar la imagen usamos view_init
#elev: es el angulo de elevación sobre el eje z
#azim: es el azimuth, rota la gráfica en los ejes x y y, es decir de manera horizontal
#Los valores por defecto utilizados son azim=-60 elev=30
ax.view_init(elev=30, azim=-60)
plt.show()
```

```python
#Python tiene 'magic functions' que nos permiten modificar un poco la manera en que la notebook funciona 
#Permite interacción con las gráficas
#SOLO FUNCIONA EN JUPYTER, NO EN COLAB
%matplotlib notebook

fig = plt.figure()
ax=Axes3D(fig)

x= np.linspace(-4,4,50)
y= np.linspace(-4,4,50)
X,Y=np.meshgrid(x,y)

ax.plot_surface(X,Y,z(X,Y), color='green')
plt.show()
```

## Duda sobre consultar columnas en un DataFrame

```python
import pandas as pd
personas = {'Nombres':['Juan','Adri','Alejandro','Pepe'], 'Edades':[21,21,20,30], 'Ciudad':['Yucatán','CDMX','Puebla','Veracruz'],'Estatura':[1.70,1.75,1.50,1.68]}
df=pd.DataFrame(personas, index=['registro1','registro2','registro3','registro4'])
print(df)
```

```python
#Para seleccionar varias columnas usamos una lista de listas dentro de df[] con el nombre de las columnas que deseamos
print(df[['Nombres','Estatura']])
```

```python
#Para seleccionar renglones por nombre 
print(df.loc[['registro1','registro3']])
```

```python
#Para seleccionar renglones con indice
print(df.iloc[[1,2]])
```

```python
#Para seleccionar columnas y renglones selectos
print(df.loc[['registro1','registro3'],['Edades','Estatura']])
```
